```python
from sklearn.datasets import make_classification  #to generate a custom classification dataset
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,random_state=41,n_clusters_per_class=1,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y,s=100,cmap='winter')
```

```python
def step(x):
  return 1 if x>=0 else 0
```

```python
def Perceptron(X,y):
  X=np.insert(X,0,1,axis=1)
  weights=np.ones(X.shape[1])
  lr=0.1
  for i in range(1000):
    j=np.random.randint(0,100)
    y_hat=step(np.dot(X[j],weights))
    weights=weights+lr*(y[j]-y_hat)*X[j]
  return weights[0],weights[1:]
```

```python
intercept_,coef_=Perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
